```python
# Create LLM
from dotenv import load_dotenv
import os
import fitz
from langchain_openai import ChatOpenAI

load_dotenv()
openrouter_key=os.getenv("OPENROUTER_API_KEY")
llm = ChatOpenAI(
    model="openai/gpt-4o-mini",
    api_key=openrouter_key,
    base_url="https://openrouter.ai/api/v1",
    temperature=0.7
)
```

```python
from langchain_core.messages import SystemMessage, HumanMessage
from qs_ranker import InstitutionQualityScorer


def my_llm(system_prompt: str, user_prompt: str) -> str:
    messages = [
        SystemMessage(content=system_prompt),
        HumanMessage(content=user_prompt)
    ]
    response = llm.invoke(messages)   # llm is your LangChain model
    return response.content

# ── Initialize once ───────────────────────────────────────────────
scorer = InstitutionQualityScorer(llm_caller=my_llm)

# ── Get just the int score ────────────────────────────────────────
score = scorer.score("International Islamic University Islamabad")
print(score)   # → 12  (int)

# ── Get full details ──────────────────────────────────────────────
result = scorer.score_full("NUST Islamabad")
print(result.score)    # → 18
print(result.tier)     # → 1
print(result.method)   # → "qs_matrix"
print(result.reason)   # → "QS 2025 rank 397 → Tier 1 | Matched: ..."

# ── Unknown university → hits LLM ─────────────────────────────────
result = scorer.score_full("Qurtuba University, D.I.Khan")  # use score_full!
print(result.score)   # → 6
print(result.reason)    # → "LLM scored" or "Both QS matrix and LLM failed — defaulted to Tier 3"
```
